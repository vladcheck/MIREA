```python
# 1.1
# Создайте класс по работе с тригонометрическими функциями. В классе
# должны быть реализованы функции вычисления:
# − косинуса;
# − синуса;
# − тангенса;
# − арксинуса;
# − арккосинуса;
# − арктангенса;
# − перевода из градусов в радианы.

import math


class TTrig:
    pass

    def sin(self, rad: float) -> float:
        return math.sin(rad)

    def cos(self, rad: float) -> float:
        return math.cos(rad)

    def tg(self, rad: float) -> float:
        return math.tan(rad)

    def asin(self, rad: float) -> float:
        return math.asin(rad)

    def acos(self, rad: float) -> float:
        return math.acos(rad)

    def atg(self, rad: float) -> float:
        return math.atan(rad)

    def deg_to_rad(self, n: float) -> float:
        return math.radians(n)


t = TTrig()
x: float = 0.5
print(f"sin({x}): {t.sin(x)}")
print(f"asin({x}): {t.asin(x)}")
print(f"cos({x}): {t.cos(x)}")
print(f"acos({x}): {t.acos(x)}")
print(f"tg({x}): {t.tg(x)}")
print(f"atg({x}): {t.atg(x)}")

degree = 30
print(f"{degree} -> {t.deg_to_rad(degree)} rad")
```

```python
from typing import Any, Self


class Tree:
    left: Self | None | float = None
    right: Self | None | float = None
    data: float

    def __init__(self, data) -> None:
        self.data = data

    def _insert(self, value: float, root: Any) -> None:
        if value < root.data:
            if root.left is None:
                root.left = Tree(value)
            else:
                self._insert(value, root.left)
        elif value > root.data:
            if root.right is None:
                root.right = Tree(value)
            else:
                self._insert(value, root.right)

    def insert_multiple(self, values: list[float]) -> None:
        for v in values:
            self._insert(v, self)

    def insert(self, value) -> None:
        self._insert(value, self)

    def PrintTree(self):
        self._PrintTree(self)

    def _PrintTree(self, root: Any) -> None:
        if root is None:
            return
        elif isinstance(root, float):
            print(root)
        else:
            self._PrintTree(root.right)
            print(root.data)
            self._PrintTree(root.left)


t: Tree = Tree(5)
t.insert_multiple([i for i in range(-4, 10)])
t.PrintTree()
```
