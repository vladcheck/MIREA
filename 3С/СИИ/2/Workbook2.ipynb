```python
"""
1.3.1
Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и 
единицами. 
"""
chessboard = np.zeros((8, 8), dtype=int)

chessboard[1::2, ::2] = 1  # Нечетные строки, четные столбцы
chessboard[::2, 1::2] = 1  # Четные строки, нечетные столбцы

print("Шахматная доска 8x8:")
print(chessboard)
```

```python
"""
1.3.2 Задание
Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания необходимо использовать функцию
"""

import numpy as np

row_values = np.arange(5)

matrix = np.zeros((5, 5))
matrix += row_values

print("Матрица 5x5 со значениями в строках от 0 до 4:")
print(matrix)
```

```python
"""
1.3.4 Задание: Создать матрицу с 0 внутри, и 1 на границах.
"""

import numpy as np

inner = np.zeros((3, 3))

matrix_with_border = np.pad(inner, pad_width=1, mode='constant', constant_values=1)

print("Матрица 5x5 с 0 внутри и 1 на границах:")
print(matrix_with_border)
```

```python
"""
2.3.3 Задание: Проделайте с получившемся из предыдущего задания фреймом данных те же действия, что и в примерах 2.2.5-2.2.7.
"""

import pandas as pd

# Загрузка данных (из предыдущего задания 2.3.2)
url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
df = pd.read_csv(url)

# 2.2.5 Анализ характеристик
print("=== Первые 5 строк (head) ===")
print(df.head())
print("\n=== Последние 3 строки (tail) ===")
print(df.tail(3))
print("\n=== Форма DataFrame (shape) ===")
print(df.shape)
print("\n=== Описательная статистика (describe) ===")
print(df.describe())

# 2.2.6 Выборка данных
print("\n=== 2.2.6: Выборка данных ===")
# Выбрать первую строку с помощью iloc
first_row = df.iloc[0]
print("Первая строка (iloc[0]):")
print(first_row)
# Выбрать столбец 'species' с помощью loc
species_column = df.loc[:, 'species']
print("\nСтолбец 'species' (loc[:, 'species']):")
print(species_column.head()) # Показываем только первые 5 для краткости

# 2.2.7 Отбор строк по условию
print("\n=== 2.2.7: Отбор строк по условию ===")
# Создаем новый DataFrame из цветков вида 'versicolor'
df_versicolor = df[df['species'] == 'versicolor']
print("DataFrame с цветками 'versicolor' (первые 5 строк):")
print(df_versicolor.head())
```

```python
"""
3.3.2 Задание: Загрузить фрейм данных по ссылке: https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака (sepal_length_cm) с использованием минимаксного преобразования, а второго
(sepal_width_cm) с задействованием z-масштабирования.
"""

import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
df = pd.read_csv(url)

df_normalized = df.copy()

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

df_normalized['sepal_length_cm'] = min_max_scaler.fit_transform(df[['sepal_length_cm']])

df_normalized['sepal_width_cm'] = standard_scaler.fit_transform(df[['sepal_width_cm']])

print("Оригинальные первые 5 строк:")
print(df.head())
print("\nНормализованные первые 5 строк:")
print(df_normalized.head())

print("\nСтатистика для нормализованного 'sepal_length_cm' (должны быть min=0, max=1):")
print(df_normalized['sepal_length_cm'].describe())
print("\nСтатистика для нормализованного 'sepal_width_cm' (должны быть mean~0, std=1):")
print(df_normalized['sepal_width_cm'].describe())
```
